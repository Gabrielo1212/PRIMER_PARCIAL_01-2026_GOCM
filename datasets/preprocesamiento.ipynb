```python
import pandas as pd
import os

CRUDOS = 'crudos'
PROCESADOS = 'procesados'
os.makedirs(PROCESADOS, exist_ok=True)

def preparar_insumos():
    print("Iniciando preparación de los 9 datasets...")

    # 1. Ames Housing
    pd.read_csv(f'{CRUDOS}/Ames Housing Dataset/archive/AmesHousing.csv').to_csv(f'{PROCESADOS}/1_ames_housing.csv', index=False)
    
    # 2. MIMIC-III (Usaremos Patients como principal para el demo)
    pd.read_csv(f'{CRUDOS}/MIMIC‑III (Medical Information Mart for Intensive Care)/mimic-iii-clinical-database-demo-1.4/PATIENTS.csv').to_csv(f'{PROCESADOS}/2_mimic_patients.csv', index=False)
    
    # 3. NHANES (Combinamos Demo y Body Measures)
    demo = pd.read_sas(f'{CRUDOS}/NHANES (National Health and Nutrition Examination Survey)/DEMO_L.xpt.txt')
    bmx = pd.read_sas(f'{CRUDOS}/NHANES (National Health and Nutrition Examination Survey)/BMX_L.xpt.txt')
    chol = pd.read_sas(f'{CRUDOS}/NHANES (National Health and Nutrition Examination Survey)/TCHOL_L.xpt.txt')
    df_temp = pd.merge(demo, bmx, on='SEQN', how='inner')

    df_final = pd.merge(df_temp, chol, on='SEQN', how='inner')

    df_final.to_csv(f'{PROCESADOS}/3_nhanes_combined.csv', index=False)
    # 4. Bike Sharing
    pd.read_csv(f'{CRUDOS}/Bike Rental (Kaggle)/bike-sharing-demand/train.csv').to_csv(f'{PROCESADOS}/4_bike_sharing.csv', index=False)

    # 5. Adult Census
    pd.read_csv(f'{CRUDOS}/Adult (Census Income) – UCI/adult/adult.data', header=None).to_csv(f'{PROCESADOS}/5_adult_census.csv', index=False)

    # 6. Credit Approval
    pd.read_csv(f'{CRUDOS}/Credit Approval – UCI/credit+approval/crx.data', header=None, na_values='?').to_csv(f'{PROCESADOS}/6_credit_approval.csv', index=False)

    # 7. Australian Credit
    pd.read_csv(f'{CRUDOS}/Statlog (Australian Credit Approval) – UCI/statlog+australian+credit+approval/australian.dat', sep=r'\s+', header=None).to_csv(f'{PROCESADOS}/7_australian_credit.csv', index=False)

    # 8. Breast Cancer
    pd.read_csv(f'{CRUDOS}/Breast Cancer Wisconsin (Original) – UCI/breast+cancer+wisconsin+original/breast-cancer-wisconsin.data', header=None).to_csv(f'{PROCESADOS}/8_breast_cancer.csv', index=False)

    # 9. Meningitis
    pd.read_csv(f'{CRUDOS}/Meningitis Dataset with Missing Values – Kaggle/mening missing 12.csv').to_csv(f'{PROCESADOS}/9_meningitis.csv', index=False)

    print("✅ ¡Los 9 insumos han sido preparados en la carpeta 'procesados'!")

preparar_insumos()
```
